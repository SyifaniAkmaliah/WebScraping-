```python
# Google Colab
# from selenium import webdriver

# Jupyter
import undetected_chromedriver as uc

import pandas as pd
import numpy as np
from datetime import datetime
import time

from selenium.common.exceptions import NoSuchElementException
```

```python
# Google Colab
# chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')
# chrome_options.add_argument('--no-sandbox')
# chrome_options.add_argument('--disable-dev-shm-usage')
# driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)

# Jupyter
driver = uc.Chrome()

driver.get('https://infocorona.serangkota.go.id/web/data')
```

```python
driver.find_element_by_xpath('/html/body/div/div/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div/label/select/option[4]').click()
```

```python
list_value = []
driver.execute_script("arguments[0].scrollIntoView();", driver.find_element_by_class_name('paginate_button.page-item.next'))

while True:
    time.sleep(1)
    
    body = driver.find_element_by_tag_name('tbody')
    row = body.find_elements_by_tag_name('tr')
    
    value = []
    
    for tr in row:
        
        for td in tr.find_elements_by_tag_name('td'):
            value.append(td.text)
            
        list_value.append(value)
        value = []
    
    try:
      driver.find_element_by_class_name('paginate_button.page-item.next.disabled')
      break
    except NoSuchElementException:
      driver.find_element_by_class_name('paginate_button.page-item.next').click()
```

```python
len(list_value)
```

```python
head = driver.find_element_by_tag_name('thead')
row = head.find_elements_by_tag_name('th')
col = []

for th in row:
    col.append(th.text)
    
col
```

```python
df = pd.DataFrame(list_value, columns=col)
df
```

```python
df.insert(1,column='provinsi', value='banten')
df.insert(2,column='kabkot', value='Serang')
df.insert(4,column='kelurahan', value='')
df.insert(11,column='types', value='kecamatan')
df.insert(12,column='types', value='https://infocorona.serangkab.go.id/data-sebaran')  
```
