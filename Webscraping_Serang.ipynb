```python
#---SCRAPING---
!pip install selenium

#install library and download webdriver
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime as dt
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
source = 'https://infocorona.serangkab.go.id/data-sebaran'
driver.get(source)
driver.find_element_by_xpath('//*[@id="v-pills-tabContent"]')
```

```python
content = driver.page_source
#print(content)
soup = BeautifulSoup(content)

#Kontak Erat
div = soup.find("table", class_="table table-hover table-border")
links=div.find_all('td')
#print(links)
#print(div)

#Kasus Suspek
div2 = soup.find("div", class_="tab-pane fade")
links2 =div2.find_all('td')
#print(div2)

#Kasus Probable
div3 = soup.find("div", class_="tab-pane fade",id="v-pills-messages")
links3 =div3.find_all('td')
#print(div3)

#Kasus terkonfirmasi
div4 = soup.find("div", class_="tab-pane fade",id="v-pills-settings")
links4 =div4.find_all('td')
```

```python
#Kontak Erat
kecamatan = []
ke_isolasi_mandiri = []
ke_selesai_pemantauan = []
ke_menjadi_kasus_suspek = []
ke_menjadi_kasus_konfirmasi = []

#Kasus Suspek
ks_dirawat_rs_rujukan = []
ks_dirawat_rs_darurat = []
ks_dirawat_isolasi_mandiri = []
ks_discarded_sembuh = []
ks_discarded_meninggal = []
ks_menjadi_kasus_konfirmasi = []

#Kasus Probable
kp_dirawat_rs_rujukan = []
kp_dirawat_rs_darurat = []
kp_meninggal = []

#Kasus Terkonfirmasi
kt_dirawat_rs_rujukan = []
kt_dirawat_rs_darurat = []
kt_dirawat_isolasi_mandiri = []
kt_sembuh = []
kt_meninggal = []

scrape_date=[]
date_update=[]
total_pdp=[]
positif_dirawat=[]
```

```python
#find time and date
date = soup.find('div',class_ = 'section_title text-center mb-95')
date1 = date.find('p')
a = date1.text[19:32].split()
bulan = ['januari', 'februari', 'maret', 'april', 'mei', 'juni','juli','agustus', 'september', 'oktober','november','desember']
bulansingkat = ['jan', 'feb', 'mar', 'apr', 'mei', 'jun','jul','august', 'sep', 'okt','nov','des']
bln = bulansingkat.index(a[1].lower())+1
#index mencari lokasi 
date_update = a[2]+'-'+str(bln)+'-'+a[0]
print(date_update)
date_update = dt.strptime(date_update, '%Y-%m-%d').strftime('%Y-%m-%d')
```

```python

#to table time scraping 
#install datetime  

now = dt.now().strftime('%Y-%m-%d')
```

```python
#append data to list
for i, link in enumerate(links):
    if i in range(6, len(links), 7):
        kecamatan.append(link.get_text().lower().capitalize())
    if i in range(7,len(links), 7):
        ke_isolasi_mandiri.append(link.get_text())
    if i in range(8,len(links), 7):
        ke_selesai_pemantauan.append(link.get_text())
    if i in range(9,len(links), 7):
        ke_menjadi_kasus_suspek.append(link.get_text())
    if i in range(10,len(links), 7):
        ke_menjadi_kasus_konfirmasi.append(link.get_text())
#print(ke_isolasi_mandiri)
for i, link in enumerate(links2):
    if i in range(10,len(links2), 8):
        ks_dirawat_rs_rujukan.append(link.get_text())
    if i in range(11,len(links2), 8):
        ks_dirawat_rs_darurat.append(link.get_text())
    if i in range(12,len(links2), 8):
        ks_dirawat_isolasi_mandiri.append(link.get_text())
    if i in range(13,len(links2), 8):
        ks_discarded_sembuh.append(link.get_text())
    if i in range(14,len(links2), 8):
        ks_discarded_meninggal.append(link.get_text())
    if i in range(15,len(links2), 8):
        ks_menjadi_kasus_konfirmasi.append(link.get_text())
#print(ks_dirawat_rs_rujukan)


for i, link in enumerate(links3):
    if i in range(6,len(links3), 5):
        kp_dirawat_rs_rujukan.append(link.get_text())
    if i in range(7,len(links3), 5):
        kp_dirawat_rs_darurat.append(link.get_text())
    if i in range(8,len(links3), 5):
        kp_meninggal.append(link.get_text())

for i, link in enumerate(links4):
    if i in range(8,len(links4), 7):
        kt_dirawat_rs_rujukan.append(link.get_text())
    if i in range(9,len(links4), 7):
        kt_dirawat_rs_darurat.append(link.get_text())
    if i in range(10,len(links4), 7):
        kt_dirawat_isolasi_mandiri.append(link.get_text())
    if i in range(11,len(links4), 7):
        kt_sembuh.append(link.get_text())
    if i in range(12,len(links4), 7):
        kt_meninggal.append(link.get_text())
    if i in range(8,len(links4), 7):
        scrape_date.append(now)
```

```python
for i in range(len(ks_dirawat_rs_rujukan)):
        q=(int(ks_dirawat_rs_rujukan[i])+int(ks_dirawat_rs_darurat[i])+int(ks_dirawat_isolasi_mandiri[i])+
           int(kp_dirawat_rs_rujukan[i])+int(kp_dirawat_rs_darurat[i]))
        total_pdp.append(q)

for i in range(len(kt_dirawat_rs_rujukan)):
        q=(int(kt_dirawat_rs_rujukan[i])+int(kt_dirawat_rs_darurat[i]))
        positif_dirawat.append(q)

```

```python
#create Tabel

df = pd.DataFrame()
```

```python
df['scrape_date'] = scrape_date
df['date_update'] = date_update        
df['kecamatan'] = kecamatan
df['total_odp'] = ke_isolasi_mandiri
df['positif_isolasi'] = kt_dirawat_isolasi_mandiri
df['positif_sembuh'] = kt_sembuh
df['positif_meninggal'] = kt_meninggal        
df['total_pdp'] = total_pdp
df['positif_dirawat'] = positif_dirawat
df.insert(3,column='kelurahan', value='')
df.insert(2,column='provinsi', value='Banten')
df.insert(3,column='kabkot', value='Serang')
df.insert(12,column='types', value='Kecamatan')
df.insert(13,column='source_link', value='https://infocorona.serangkab.go.id/data-sebaran')  
```
